### Obtaining list of prices from the website 

```python
prices = soup.find_all("bdi")
prices
```

```python
len(prices)
```

```python
priceList=[]

for price in prices:
    pricy = float(price.text[1:])
    soapprice = pricy
    priceList.append(soapprice)
```

```python
priceList
```

#### The price of 0.00 can be ignored and also 27.00 as this is the former price.

```python
len(priceList)
```

```python
list1= priceList[slice(1,-3)]
list1
```

```python
list2= priceList[slice(16,18)]
list2
```

```python
SoapPrices = list1 + list2
```

```python
SoapPrices
```

```python
len(SoapPrices)
```

```python
Price=[]

for soap in SoapPrices:
    soapInfo = {"price":soap}
    Price.append(soapInfo)
```

```python
Price
```

```python
SoapName
```

```python
zipped = zip(SoapName, Price)
zipped
```

```python
soapInfo=[{**name, **price} for name, price in zipped]
soapInfo
```
